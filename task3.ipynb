```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('householdtask3.csv')
```

```python
data.head(5)
```

```python
# Bar chart
plt.bar(data['year'],data['size'])

# Title of Chart
plt.title('BAR Chart')

# Labeling of Axis
plt.xlabel('YEAR')
plt.ylabel('SIZE')

# Plotting Legend
plt.show()

```

```python
# Line Chart
plt.plot(data['year'])
plt.plot(data['income'])

# Title of Chart
plt.title('Line Chart')

# Labeling of Axis
plt.xlabel('YEAR')
plt.ylabel('INCOME')

# Plotting Legend
plt.show()
```
